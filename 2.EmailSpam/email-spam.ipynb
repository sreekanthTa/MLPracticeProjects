```python
# 1.Load Data 
import pandas as pd

data  = pd.read_csv("./emails.csv")
# data.shape
data.head()
```

```python
# Split data
x = data.drop(columns=["Prediction","Email No."], axis=1)
y = data["Prediction"]
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# 2.Load Model 
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf
```

```python
# 3. Fit data into the Model 
clf.fit(X=x_train,y=y_train)
```

```python
train_score = clf.score(X=x_train, y=y_train)
train_score * 100
```

```python
y_predict = clf.predict(X=x_test)
```

```python
test_score = clf.score(X=x_test,y=y_test)
test_score * 100
```

```python
# CHECK METRICS
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_predict))
```

```python
print(confusion_matrix(y_test, y_predict))
```

```python
print(accuracy_score(y_test, y_predict))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'n_estimators': [i for i in range(100, 200, 10)]}

grid = GridSearchCV(estimator=RandomForestClassifier(), param_grid=param_grid, cv=5, verbose=1)

grid.fit(x, y)
```

```python
# Find the best parameters
print(f"The best parameter values are: {grid.best_params_}")
print(f"With a score of: {grid.best_score_*100:.2f}%")
```

```python
clf =  grid.best_estimator_
clf.fit(x_train, y_train)
```

```python
y_predict = clf.predict(X=x_test)

print(accuracy_score(y_test, y_predict))
```
